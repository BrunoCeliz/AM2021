```python
import pandas as pd #librería para datos
import matplotlib.pyplot as plt #para el hist
from scipy import stats as st
import numpy as np

D = pd.read_csv('train.csv') #DF de entrenamiento. Acá sabemos el H_class de cada ID (sabemos qué es cada objeto)
D
```

```python
D.drop('DEdeg',axis=1).drop('RAdeg',axis=1) #Tiro RA y DEC
```

```python
#Armamos varios IC:
```

```python
#Long de onda pequeña - grande, contigua (pero podemos probar difs):
D['H3-K3'] = D['H3mag'] - D['K3mag']
D['J3-H3'] = D['J3mag'] - D['H3mag']
D['Y3-J3'] = D['Y3mag'] - D['J3mag']
D['Z3-Y3'] = D['Z3mag'] - D['Y3mag']

D['Y-J'] = D['Ymag'] - D['Jmag']
D['J-H'] = D['Jmag'] - D['Hmag']

D['i-z'] = D['imag'] - D['zmag']
D['r-i'] = D['rmag'] - D['imag']
D['g-r'] = D['gmag'] - D['rmag']
D['u-g'] = D['umag'] - D['gmag']
```

```python
#Tiramos las mags (de a 1 pls):
```

```python
D.drop('Kmag',axis=1,inplace=True)
```

```python
D['W1-W2'] = D['W1mag'] - D['W2mag']
D['u3-g3'] = D['u3mag'] - D['g3mag']
D['g3-r3'] = D['g3mag'] - D['r3mag']
```

```python
D.drop('i3mag',axis=1,inplace=True)
D
```

```python
D.describe()
```

```python
#Copio y pego de Try1 para SVM:
```

```python
#Separo la columna Hclass del resto, y la guado aparte, como para hacer un 'test' de tu 'train':

X = D[D.columns.drop(['Hclass','ID'])]; y = D['Hclass']; ID = D['ID']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y) #para separar en 75% tr - 25% test de mi conjt train og.
```

```python
y.value_counts()
```

```python
from sklearn.svm import LinearSVC, SVC
from sklearn.metrics import f1_score

qualy = LinearSVC(max_iter=5000,class_weight='balanced') #def mi clasificador
qualy.fit(X_train,y_train) #entreno
```

```python
y_true, y_predic = y_train, qualy.predict(X_train) #def la posta y lo que me devuelve como ML del test inventado
f1_score(y_true,y_predic,average=None)
```

```python
y_true, y_predic = y_train, qualy.predict(X_train) #def la posta y lo que me devuelve como ML del test inventado
f1_score(y_true,y_predic,average='weighted') 
```

```python
y_true, y_predic = y_test, qualy.predict(X_test) #def la posta y lo que me devuelve como ML del test inventado
f1_score(y_true,y_predic,average='weighted')
```
